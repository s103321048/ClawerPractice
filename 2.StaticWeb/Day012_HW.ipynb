### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

```python
'''
Your Code
'''
import requests
from bs4 import BeautifulSoup
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html")
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
'''
Your Code
'''
cat_dict = {}
for d in soup.find(class_="part_list_2").find_all('h3'):
    cat_name = d.find('em').text
    if cat_name not in cat_dict:
        cat_dict[cat_name] = [d.find_all('a')[-1].text]
    else:
        cat_dict[cat_name].append(d.find_all('a')[-1].text)
cat_dict
```

### ③ 哪一個類別的文章最多

```python
'''
Your Code
'''
for i in cat_dict:
    print(i, "\t" ,len(cat_dict[i]))
```
