### 1. 取出知乎問題發問時間

```python
import requests
r = requests.get("https://www.zhihu.com/api/v4/questions/55493026/answers")
r.text
# 這是直接抓，沒有家 header，會回傳 400 Bad Request
```

```python
header = {
#     "authority": "www.zhihu.com",
#     "method": "GET",
#     "path": "/api/v4/questions/55493026/answers",
#     "scheme": "https",
#     "accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9",
#     "accept-language": "en-US,en;q=0.9,zh-TW;q=0.8,zh;q=0.7",
#     'sec-fetch-dest': "document",
#     'sec-fetch-mode': "navigate",
#     'sec-fetch-site': "none",
    'user-agent': "user"
}

r1 = requests.get("https://www.zhihu.com/api/v4/questions/55493026/answers", headers=header)
r1.text
```

```python
# 1. 取出知乎問題發問時間

'''
Your Code
'''
import json
info = json.loads(r1.text)
info
```

```python
for i in info['data']:
    print(i['updated_time'])
```

### 2. 取出第一筆與最後一筆回答的時間

```python
# 2. 取出第一筆與最後一筆回答的時間

'''
Your Code
'''
print('start:', info['data'][0]['updated_time'])
print('end:', info['data'][-1]['updated_time'])
```
